### 2.10. Check Outliers

```python
# Box plots to identify outliers
def plot_boxplots(data):
    numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns
    plt.figure(figsize=(20, 15))
    for i, col in enumerate(numerical_cols, 1):
        plt.subplot(4, 3, i)
        sns.boxplot(x=data[col])
        plt.title(f'Box plot of {col}')
    plt.tight_layout()
    plt.show()

# Plotting box plots for outliers
plot_boxplots(data)
```

```python
# Function to calculate the number of outliers in each column and return as DataFrame
def count_outliers_df(data):
    outlier_counts = {'Column': [], 'Outlier Count': []}
    
    for col in data.select_dtypes(include=['float64', 'int64']).columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Count outliers
        outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
        outlier_counts['Column'].append(col)
        outlier_counts['Outlier Count'].append(len(outliers))
        
    return pd.DataFrame(outlier_counts)

# Calculate the number of outliers for each numerical column and convert to DataFrame
outlier_counts_df = count_outliers_df(data)

# Show the DataFrame
outlier_counts_df
```

3.2.1. MOV (Median Order Value)
- Given the skewed distribution in the box plots, using the median instead of the average (mean) might be more appropriate, especially for calculating the Average Order Value (AOV).
- Reasons:

| **Reason** | **Interpretation**                                        |
|:---------------:|:-------------------------------------------------------------|
| Robustness to Outliers     | The median is less affected by extreme values or outliers. Since the distribution is skewed, the mean could be disproportionately influenced by a few very high or very low values, leading to a potentially misleading representation of the central tendency.    |
|  Better Representation of Central Tendency     | In skewed distributions, the median provides a better sense of the “typical” value, as it represents the middle point of the data, where 50% of the values are below and 50% are above.     |
| Skewed Data Consideration      | In the presence of skewness, the mean tends to be pulled in the direction of the skew, while the median remains a more reliable measure of central tendency.   |

3.5.1. MOV of Discounted And Non-Discounted Orders By Payment Method

```python
# Calculate MOV by Payment Method
mov_by_payment_discounted = discounted_orders.groupby('Payment_Method')['Order_Total'].median()
mov_by_payment_non_discounted = non_discounted_orders.groupby('Payment_Method')['Order_Total'].median()
```

```python
# Create a DataFrame to compare MOV by payment method between discounted and non-discounted orders
mov_by_payment = pd.DataFrame({
    'Discounted_Orders_MOV': mov_by_payment_discounted,
    'Non_Discounted_Orders_MOV': mov_by_payment_non_discounted
})

# Fill NaN values with 0 to handle any missing payment methods in either category
mov_by_payment.fillna(0, inplace=True)

# Format the MOV columns with Rupee symbol
mov_by_payment['Discounted_Orders_MOV'] = mov_by_payment['Discounted_Orders_MOV'].apply(lambda x: f"₹ {x:,.2f}")
mov_by_payment['Non_Discounted_Orders_MOV'] = mov_by_payment['Non_Discounted_Orders_MOV'].apply(lambda x: f"₹ {x:,.2f}")

# Calculate the total MOV for each category (without formatting for accurate calculation)
total_mov_discounted = mov_by_payment_discounted.sum()
total_mov_non_discounted = mov_by_payment_non_discounted.sum()

# Add percentage columns
mov_by_payment['Discounted_Orders_Percentage'] = (mov_by_payment_discounted / total_mov_discounted) * 100
mov_by_payment['Non_Discounted_Orders_Percentage'] = (mov_by_payment_non_discounted / total_mov_non_discounted) * 100

# Format the percentage columns as strings with a percentage sign
mov_by_payment['Discounted_Orders_Percentage'] = mov_by_payment['Discounted_Orders_Percentage'].map('{:.2f}%'.format)
mov_by_payment['Non_Discounted_Orders_Percentage'] = mov_by_payment['Non_Discounted_Orders_Percentage'].map('{:.2f}%'.format)

# Display the DataFrame
mov_by_payment
```

3.8.2. Statistical Proof Of Correlation Between Sales And Discount Trend Lines

```python
# Extract relevant data columns
sales = monthly_sales['Order_Total']
discounts = monthly_sales['Discount_Amount']

# Calculate the Spearman rank correlation coefficient and the p-value
spearman_correlation, p_value = stats.spearmanr(discounts, sales)

print(f"Spearman Rank Correlation Coefficient: {spearman_correlation}")
print(f"P-value: {p_value}")

# Interpretation of results
if p_value < 0.05:
    print("There is a statistically significant correlation between discounts and sales.")
else:
    print("There is no statistically significant correlation between discounts and sales.")
```

```python
# Calculate the Kendall Tau correlation coefficient and the p-value
kendall_tau, p_value = stats.kendalltau(discounts, sales)

print(f"Kendall's Tau Correlation Coefficient: {kendall_tau}")
print(f"P-value: {p_value}")

# Interpretation of results
if p_value < 0.05:
    print("There is a statistically significant correlation between discounts and sales.")
else:
    print("There is no statistically significant correlation between discounts and sales.")
```

# 4. Data Pre-processing (Multiclass)
Pada tahap ini, data akan disiapkan untuk pemodelan Machine Learning. Karena fokus analisis adalah untuk melihat status transaksi secara keseluruhan, menggunakan pendekatan klasifikasi multikelas (Multiclass Classification) dengan target: `complete`, `canceled`, dan `refunded`.

### 4.1. Filtering & Sampling Data
Filter dataset hanya untuk 3 status target. Mengingat dataset sangat besar, mengambil sampel 15.000 data secara acak untuk mencegah *bottleneck* komputasi pada algoritma yang berat seperti SVM.

```python
# Memfilter data untuk 3 kelas yang penulisan namanya tepat
data_ml = data[data['status'].isin(['complete', 'canceled', 'order_refunded'])].copy()

# Mengambil sampel acak 15.000 data
data_ml = data_ml.sample(n=15000, random_state=42)
print(f"Total data setelah disampel: {data_ml.shape[0]} baris.")
```

### 4.2. Pemilihan Fitur & Encoding
Memilih fitur prediktor (numerik & kategorikal). Target status akan diubah menjadi angka menggunakan `LabelEncoder`, sedangkan fitur kategorikal akan diubah menggunakan metode `One-Hot Encoding`.

```python
# 1. Menentukan fitur prediktor (X) dan target (y)
# MENGGUNAKAN NAMA KOLOM BARU YANG SUDAH DI-RENAME
features = ['Price', 'Quantity_Ordered', 'Order_Total', 'Product_Category', 'Payment_Method']
X = data_ml[features]

# Kolom status tetap huruf kecil karena tidak ikut di-rename
y = data_ml['status'] 

# 2. Label Encoding untuk Target
le_target = LabelEncoder()
y_encoded = le_target.fit_transform(y)

# Mencetak mapping kelas
class_mapping = dict(zip(le_target.classes_, le_target.transform(le_target.classes_)))
print(f"Mapping Target: {class_mapping}")

# 3. One-Hot Encoding untuk fitur kategorikal
# MENGGUNAKAN NAMA KOLOM BARU UNTUK KATEGORI DAN PEMBAYARAN
X_encoded = pd.get_dummies(X, columns=['Product_Category', 'Payment_Method'], drop_first=True)

print(f"Proses Encoding Selesai! Dimensi X_encoded: {X_encoded.shape}")
```

### 4.3. Train-Test Split
Membagi data yang sudah di-*encode* menjadi data latih (80%) dan data uji (20%).

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

print(f"Dimensi X_train: {X_train.shape}")
print(f"Dimensi X_test: {X_test.shape}")

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

print(f"Dimensi X_train: {X_train.shape}")
print(f"Dimensi X_test: {X_test.shape}")
```

# 5. Machine Learning & Ablation Study
Pada bagian ini, melatih 4 algoritma (*Logistic Regression, Naive Bayes, Random Forest, SVM*) dalam 4 skenario pemrosesan data yang berbeda (*Ablation Study*) untuk melihat algoritma mana dan pendekatan *preprocessing* apa yang memberikan hasil terbaik.

### 5.1. Inisialisasi Model & Setup
Menyiapkan model dan membuat fungsi iterasi agar evaluasi lebih efisien. Metrik yang digunakan adalah *Accuracy* dan *Weighted F1-Score* yang cocok untuk *Multiclass*.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score
import time

# Siapin list kosong buat nampung semua hasil evaluasi nanti buat dibikin tabel di akhir
results = []

# Daftar model yang mau diuji (SVM udah dikasih max_iter=2000 biar nggak looping 24 jam lagi!)
models = {
    'Logistic Regression': LogisticRegression(max_iter=2000, random_state=42),
    'Naive Bayes': GaussianNB(),
    'Random Forest': RandomForestClassifier(random_state=42, n_estimators=50),
    'SVM': SVC(kernel='linear', random_state=42, max_iter=2000) 
}

# Bikin fungsi evaluasi otomatis biar nggak usah ngetik kode berulang-ulang buat tiap skenario
def evaluate_models(X_train_data, X_test_data, y_train_data, y_test_data, scenario_name):
    for name, model in models.items():
        start_time = time.time()
        model.fit(X_train_data, y_train_data)
        y_pred = model.predict(X_test_data)
        end_time = time.time()
        
        acc = accuracy_score(y_test_data, y_pred)
        f1 = f1_score(y_test_data, y_pred, average='weighted')
        
        results.append({
            'Scenario': scenario_name,
            'Model': name,
            'Accuracy': acc,
            'F1-Score': f1,
            'Time (s)': round(end_time - start_time, 2)
        })
        
print("Setup Model & Fungsi Evaluasi berhasil dimuat!")
```

### 5.2. Skenario 1: Baseline
Melatih model menggunakan data mentah hasil *encoding* tanpa perlakuan tambahan.

```python
evaluate_models(X_train, X_test, y_train, y_test, '1. Baseline')
print("Skenario 1 selesai dijalankan.")
```

### 5.3. Skenario 2: Feature Selection
Mereduksi dimensi dengan memilih 10 fitur terbaik yang paling berkorelasi dengan target menggunakan metode `SelectKBest`.

```python
# Tambahkan baris import ini agar Python mengenali SelectKBest dan f_classif
from sklearn.feature_selection import SelectKBest, f_classif

k_best = min(10, X_train.shape[1])
selector = SelectKBest(score_func=f_classif, k=k_best)
X_train_fs = selector.fit_transform(X_train, y_train)
X_test_fs = selector.transform(X_test)

evaluate_models(X_train_fs, X_test_fs, y_train, y_test, '2. Feature Selection')
print("Skenario 2 selesai dijalankan.")
```

### 5.4. Skenario 3: Standardization
Menyamakan skala angka pada semua fitur menggunakan `StandardScaler`. Hal ini krusial untuk algoritma berbasis jarak/skala numerik.

```python
from sklearn.preprocessing import StandardScaler

# Skenario 3: buktikan apakah menstandarisasi skala data bisa menyembuhkan "penyakit" lama SVM dan LogReg.
# Ingat: fit_transform HANYA untuk data train, test cuma boleh di-transform biar nggak ada data leakage.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

evaluate_models(X_train_scaled, X_test_scaled, y_train, y_test, '3. Scaled Data')
print("Skenario 3 selesai dijalankan.")
```

### 5.5. Skenario 4: Dimensionality Reduction (PCA)
Menggunakan *Principal Component Analysis* (PCA) untuk mereduksi seluruh fitur menjadi 5 komponen utama. PCA dilakukan pada data yang sudah di-*scale* (Skenario 3).

```python
from sklearn.decomposition import PCA

# Skenario 4: Kompres semua fitur pakai PCA.
# Targetnya: simpan 95% informasi (variance) utamanya.
pca = PCA(n_components=0.95, random_state=42)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

evaluate_models(X_train_pca, X_test_pca, y_train, y_test, '4. Scaled + PCA')
print("Skenario 4 selesai dijalankan.")
```

### 5.6. Hasil Evaluasi & Komparasi
Menampilkan tabel komparasi performa dari setiap model pada keempat skenario *Ablation Study* yang telah dilakukan.

```python
import pandas as pd
from IPython.display import display

# Kumpulin semua list 'results' ke dalam bentuk DataFrame pandas biar enak dilihat dan dibandingin.
results_df = pd.DataFrame(results)

# Urutkan berdasarkan Skenario, lalu urutkan Akurasi dari yang paling tinggi di tiap skenario
results_df = results_df.sort_values(by=['Scenario', 'Accuracy'], ascending=[True, False]).reset_index(drop=True)

print("=== HASIL KESELURUHAN ABLATION STUDY ===")
display(results_df)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
import time

# 1. Inisialisasi "mesin" kosong
logreg = LogisticRegression(random_state=42, max_iter=1000)

# 2. Tentukan daftar "onderdil" (parameter) yang mau dicoba-coba
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100], # Kekuatan regulariasi
    'penalty': ['l2'] # L2 (Ridge) stabil untuk solver default
}

# 3. Setup pencarian otomatis (Cross-Validation 5 lipatan)
grid_search = GridSearchCV(logreg, param_grid, cv=5, scoring='accuracy', n_jobs=-1)

print("Mulai mencari setelan terbaik... (tunggu sebentar ya ⏳)")
start_time = time.time()

# 4. Latih model dengan data yang SUDAH DI-SCALE (Skenario 3)
grid_search.fit(X_train_scaled, y_train)
end_time = time.time()

# 5. Simpan model pemenangnya!
best_logreg = grid_search.best_estimator_

print(f"Pencarian selesai dalam {round(end_time - start_time, 2)} detik! 🎉")
print(f"Setelan Parameter Terbaik: {grid_search.best_params_}")
print(f"Akurasi Validasi (Cross-Val): {grid_search.best_score_:.4f}")
```

### 6.3. Evaluasi Mendalam (Confusion Matrix & Classification Report)

Angka akurasi saja terkadang bisa menipu, terutama jika jumlah data per kelasnya tidak seimbang (*imbalanced*). Untuk melihat "kualitas" tebakan model kita, kita perlu membedahnya menggunakan *Confusion Matrix*.

Dari matriks ini, kita bisa melihat: **Ketika model salah menebak, dia salahnya ke mana?** Apakah pesanan yang sebenarnya *canceled* malah ditebak *complete*? Mari kita lihat visualisasinya.

```python
from sklearn.metrics import confusion_matrix, classification_report

# 1. Gunakan model terbaik hasil tuning tadi untuk memprediksi data ujian (X_test)
y_pred_best = best_logreg.predict(X_test_scaled)

# Catatan: Berdasarkan proses encoding sebelumnya, labelnya adalah:
# 0: canceled | 1: complete | 2: order_refunded
target_labels = ['canceled (0)', 'complete (1)', 'order_refunded (2)']

# 2. Cetak Rapor Detailnya
print("=== CLASSIFICATION REPORT ===")
print(classification_report(y_test, y_pred_best, target_names=target_labels))

# 3. Bikin Visualisasi Confusion Matrix
cm = confusion_matrix(y_test, y_pred_best)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=target_labels, yticklabels=target_labels)
plt.title('Confusion Matrix - Best Logistic Regression', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Tebakan Model (Predicted Label)', fontsize=12, labelpad=10)
plt.ylabel('Kenyataan Asli (True Label)', fontsize=12, labelpad=10)
plt.tight_layout()
plt.show()
```

Ini merupakan indikator yang sangat positif bagi proses operasional, karena memungkinkan adanya deteksi dini terhadap pesanan-pesanan yang memiliki probabilitas tinggi untuk dibatalkan.
* **Limitasi Model (Order Refunded):** Titik kelemahan paling signifikan dari model ini terletak pada prediksi kelas `order_refunded`. ### 6.4. Feature Importance (Mencari Tahu "Why?")

Ini adalah bagian paling berharga bagi tim Bisnis/Marketing. Model tidak hanya bisa menebak, tapi juga bisa memberitahu kita **kolom/fitur apa yang paling memengaruhi keputusan pelanggan**. 

Dengan mengekstrak koefisien dari model Logistic Regression, kita bisa melihat 10 faktor utama (*Top 10 Features*) yang menyetir pola transaksi di *dataset* ini.

```python
import numpy as np

# 1. Ambil nama-nama kolom dari X_encoded (sebelum di-split, agar namanya terbaca)
feature_names = X_encoded.columns

# 2. Ambil bobot (koefisien) dari model terbaik
# Karena target kita ada 3 kelas, kita ambil rata-rata pengaruh absolut dari setiap fitur ke semua kelas
importance = np.mean(np.abs(best_logreg.coef_), axis=0)

# 3. Masukkan ke dalam tabel (DataFrame) biar rapi
feat_imp_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importance
}).sort_values(by='Importance', ascending=False)

# 4. Ambil 10 fitur teratas saja
top_10_features = feat_imp_df.head(10)

# 5. Visualisasikan!
plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_features, x='Importance', y='Feature', palette='magma')

plt.title('Top 10 Fitur Paling Berpengaruh dalam Prediksi Status Pesanan', fontsize=15, fontweight='bold', pad=15)
plt.xlabel('Tingkat Pengaruh (Absolute Coefficient)', fontsize=12, labelpad=10)
plt.ylabel('Nama Fitur', fontsize=12, labelpad=10)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Berdasarkan visualisasi *Top 10 Feature Importance*, dapat diidentifikasi variabel-variabel prediktor utama yang paling signifikan dalam memengaruhi probabilitas transisi status pesanan pelanggan.
